```python
# #行业：应用

# 融资状态：未知

# 地区：北京市 朝阳区

# 公司规模：50-150人

# 网址：https://micous.com
tile_list = ["name","行业","融资状态","地区","公司规模","网址"]
data = {}
for item in tile_list:
    data[item] = []
import requests
from bs4 import BeautifulSoup
def getdata(url):
    r = requests.get(url,timeout=8)
    r.encoding='utf-8'
    soup = BeautifulSoup(r.text)
    li_list = soup.findAll(name="li", attrs={"class" :"cru-list"})
    for li_item in li_list:
        div_list = li_item.findAll(name="div", attrs={"class" :"crr-l"})
        p_list = div_list[0].findAll(name="p")
        p_list = div_list[0].findAll(name="p")
        for i,x in enumerate(p_list):
            data[tile_list[i+1]].append(x.text.split("：")[1])
        name = li_item.find(name="h2").a.text
        data[tile_list[0]].append(name)
number = 859
#number = 10
for i in range(number):
    url = "https://www.baijingapp.com/company/page-%s"%i
    try:
        getdata(url)
    except Exception as e:
        print("error "+ str(i))
    print(i)
```

```python
import  pandas as pd
pd_data = pd.DataFrame(data)
writer=pd.ExcelWriter('data_baijingapp.xlsx',engine='xlsxwriter')
pd_data.to_excel(writer,index=False)
writer.save()
```
