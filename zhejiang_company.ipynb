```python
import requests
from bs4 import BeautifulSoup
import json
urlList = []
for i in range(1,7):
    urlbase = "http://www.zsia.org/page/{}?typeCode=254117&rows=15".format(i)   #http://www.zsia.org/hyzc/lay-254117.html
    r = requests.get(urlbase)
    data = r.json()
    # 将JSON数据转换为字典
    for item in data['rows']:
        urlList.append(item['id'])
print(urlList)
```

```python
comapany_dict = {}
for id in urlList:  #http://www.zsia.org/hyzc/4315.html
    try:
        url = "http://www.zsia.org/hyzc/{}.html".format(id)
        r = requests.get(url)
        r.encoding='utf-8'
        soup = BeautifulSoup(r.text)
        r = requests.get(url)
        r.encoding='utf-8'
        soup = BeautifulSoup(r.text)

        contents = soup.find(name="div", attrs={"class" :"content"})
        company = ""
        count = 0
        for item_p in contents:
            cont = item_p.text
            if len(cont) > 1:
                if cont[-2:] == '公司':
                    company = cont
                elif company != "" :
                    comapany_dict[company] = cont
                    company = ""
                    count = count + 1
        print("{} over nums {}".format(id,count))
    except Exception as e:
        print("{} error nums".format(id))
   
```

```python
import pandas as pd
writer=pd.ExcelWriter('zhejiang_companys.xlsx',engine='xlsxwriter')
dfOneSch = pd.DataFrame({'公司名称':comapany_dict.keys(),'简介':comapany_dict.values()})
dfOneSch.to_excel(writer,index=False)
writer.save()
```
