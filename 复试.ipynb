```python
mat = [ [ 0 for x in range(10)] for y in range(10)]

dirction = [[-1,0],[0,1],[1,0],[0,-1]]
food = [6,6]
point = [[2,2]]
snk = [[[2,2],[2,1]]]
init_snake = [[2,2]]
mat[init_snake[0][0]][init_snake[0][1]] = 1
def setWall(snake,matw):
    for item in snake:
        matw[item[0]][item[1]] = -1
def goOne(x,y,snake,head):
    matw = [ [ 0 for x in range(10)] for y in range(10)]
    setWall(snake,matw)
    global mat
    if x>=0 and x<=9 and y>=0 and y<=9 and matw[x][y] != -1 and (mat[x][y] == 0 or mat[x][y]>step):
        mat[x][y] = step
        return True
    return False
def nextSnake(snake,d):
    head = [snake[0][0] + d[0],snake[0][1] + d[1]]
    return [head + snake[:-1]]
def nextStep(point,snk,step):
    ppoint = []
    ssnk = []
    for x in range(len(point)):
        snake = snk[x]
        head = snake[0]
        for d in dirction:
            x = head[0] + d[0]
            y = head[1] + d[1]
            if goOne(x,y,snake,step):
                ppoint.append([x,y])
                newsnake = nextSnake(snake,d)
                ssnk.append(newsnake)
    return ppoint,ssnk

for step in range(2,100):
    point,snk = nextStep(point,snk,step)
    if mat[food[0]][food[1]] != 0:
        break
def getDirection():
    head = food
    x,y = 0,0
    step = mat[food[0]][food[1]]
    while True:
        if step == 2:
            break
        for d in dirction:
            x = head[0] + d[0]
            y = head[1] + d[1]
            if mat[x][y] == step - 1:
                head = x,y
                step = step -1
                break
    return x,y
def judge(snake,nextsnake):
    if nextsnake[0]>snake[0]:
        print("down")
    elif nextsnake[0]<snake[0]:
        print("up")
    elif nextsnake[1]>snake[1]:
        print("right")
    elif nextsnake[1]<snake[1]:
        print("left")

x,y = getDirection()
judge(init_snake[0],(x,y))
```
